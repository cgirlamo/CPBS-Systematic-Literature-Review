```python
import pandas as pd
```

```python
df = pd.read_excel("Boolean_keywords.xlsx",sheet_name="Keywords")
ndf = pd.DataFrame(columns=df.columns)
ndf['Concept Group'] = df['Concept Group']
ndf.head()
```

```python
def make_bool(string):
    nstring = "("
    l = string.split(",")
    for x in l:
        if x is not l[-1]:
            nstring += (f'"{x.strip(' ')}" OR ')
        else:
            nstring += f'"{x.strip(' ')}")'
    nstring += ")"
    return(nstring)
```

```python
for val in df.index:
    for col in df.columns:
        if (val == 0) and (col == "Core Terms (used in all domains)"):
            ndf.loc[val,col] = make_bool(df.loc[val,col])
        elif (val != 1) and (col == "Core Terms (used in all domains)"):
            x = ndf.iloc[0,1] + f" AND {make_bool(df.loc[val,col])}"
            ndf.loc[val,col] = x
        elif (col != "Core Terms (used in all domains)") and (col != "Concept Group") and (val != 0):
            x = ndf.iloc[0,1] + f"AND {df.loc[val,"Core Terms (used in all domains)"]} AND {make_bool(df.loc[val,col])}"
            ndf.loc[val,col] = x
            
        
print(ndf)
```
